### representational symmetry in anatomical hippocampus

```python
symm_ab_hip = pd.read_csv("aggregated_AB_lat_hip_subregions_symmetry.csv")
symm_ac_hip = pd.read_csv("aggregated_AC_lat_hip_subregions_symmetry.csv")

def make_asym(df, fwd_label, bwd_label):
    fwd = df[df['comparison'] == fwd_label].copy()
    bwd = df[df['comparison'] == bwd_label].copy()

    fwd['forward'] = fwd['within_sim'].values
    fwd['backward'] = bwd['within_sim'].values
    fwd['asymmetry'] = fwd['forward'] - fwd['backward']

    return fwd

ab_hip = make_asym(symm_ab_hip, 'AB', 'BA')
ac_hip = make_asym(symm_ac_hip, 'AC', 'CA')

symm_df = pd.concat([ab_hip, ac_hip], ignore_index=True)
symm_df['triplet'] = symm_df['triplet'].astype(int)
```
